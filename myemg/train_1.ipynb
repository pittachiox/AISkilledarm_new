```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------
# 1. Load CSV
# ----------------------
data = pd.read_csv("emg_log1.csv").rename(columns={"Raw": "digital", "Voltage": "volt"})


# ----------------------
# 2. Preprocessing
# ----------------------
data["digital_norm"] = data["digital"] / 4095.0

# Moving Average Filter
window = 50
data["digital_filt"] = data["digital_norm"].rolling(window, center=True).mean()

# ----------------------
# 3. Threshold-based Labeling
# ----------------------
labels = []
for val in data["digital"]:
    if val > 2500:
        labels.append("grip")       # กำมือสุด ๆ
    elif val < 500:
        labels.append("open")       # แบมือ
    elif 1000 <= val <= 2000:
        labels.append("hold")       # กำค้าง
    else:
        labels.append("transition") # ช่วงเปลี่ยน

data["label"] = labels

# ----------------------
# 4. Visualization
# ----------------------
plt.figure(figsize=(12,6))
plt.plot(data["digital"], label="Raw Digital")
plt.plot(data["digital_filt"]*4095, label="Filtered (scaled)")
plt.title("EMG Digital Signal with Grip/Open States")
plt.xlabel("Sample")
plt.ylabel("ADC value")
plt.legend()
plt.show()

# Show label distribution
print(data["label"].value_counts())
```

```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, LSTM, Dense, MaxPooling1D, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# 1. การเตรียมข้อมูล

# โหลดข้อมูลจากไฟล์ emg_log3.csv
df = pd.read_csv('emg_log3.csv')
emg_signal = df['Voltage'].values

# กำหนดขนาดของ window และการเลื่อน window
window_size = 200  # จำนวนจุดข้อมูลในแต่ละ window
step_size = 50     # จำนวนจุดข้อมูลที่เลื่อนไปในแต่ละครั้ง

# ทำการ windowing และสร้าง label เบื้องต้น (ด้วยเกณฑ์ threshold-based)
X_list = []
y_list = []
threshold = 0.5  # กำหนดค่าเกณฑ์ที่ใช้ในการแยกคลาส (ปรับค่าได้ตามความเหมาะสม)

for i in range(0, len(emg_signal) - window_size, step_size):
    window = emg_signal[i:i + window_size]
    # ปรับรูปทรงให้เป็น (window_size, 1) เพื่อป้อนเข้า CNN
    X_list.append(window.reshape(-1, 1)) 
    
    # ติดป้ายกำกับตามค่าเฉลี่ยของ window
    if np.mean(np.abs(window)) > threshold:
        y_list.append('active')
    else:
        y_list.append('rest')

# แปลง list ให้เป็น numpy array
X = np.array(X_list)
y = np.array(y_list)

print(f'Shape ของ X (windowed data): {X.shape}')
print(f'Shape ของ y (labels): {y.shape}')

# ตรวจสอบจำนวนข้อมูลในแต่ละคลาส
unique, counts = np.unique(y, return_counts=True)
print(dict(zip(unique, counts)))

# 2. สร้างโมเดลและฝึก

# Encode label
le = LabelEncoder()
y_enc = le.fit_transform(y)
y_cat = to_categorical(y_enc)

# แบ่งข้อมูลเป็นชุดฝึกและชุดทดสอบ
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.2, random_state=42)

# สร้างโมเดล CNN-LSTM
model = Sequential([
    Conv1D(filters=64, kernel_size=5, activation='relu', input_shape=(window_size, 1)),
    MaxPooling1D(pool_size=2),
    Dropout(0.25),
    LSTM(units=100, return_sequences=False),
    Dense(units=len(np.unique(y)), activation='softmax')
])

# Compile model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# ฝึกโมเดล
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.1, verbose=1)

# 3. การประเมินผล
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f'ความแม่นยำบนข้อมูลทดสอบ: {accuracy*100:.2f}%')

# พล็อตความแม่นยำ
plt.figure(figsize=(12, 6))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 1. การโหลดข้อมูลและพล็อตเพื่อการติดป้ายกำกับด้วยมือ
df = pd.read_csv('emg_log3.csv')
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df.set_index('Timestamp', inplace=True)

# พล็อตข้อมูลเพื่อให้คุณเห็นภาพรวม
plt.figure(figsize=(15, 6))
plt.plot(df.index, df['Raw'], label='Raw Raw Signal', color='skyblue')
plt.title('EMG Signal from emg_log3.csv')
plt.xlabel('Timestamp')
plt.ylabel('Raw')
plt.grid(True)
plt.legend()
plt.show()



# 2. การสร้างป้ายกำกับด้วยมือ (Manual Labeling)
# ************** ให้คุณแก้ไขช่วงเวลาด้านล่างนี้ตามการสังเกตของคุณเอง **************
# กำหนดป้ายกำกับเริ่มต้นทั้งหมดเป็น 'unknown'
df['Label'] = 'unknown'

# ตัวอย่างการกำหนดป้ายกำกับด้วยมือ:
# กำหนดช่วงเวลาที่ 'active'
active_intervals = [
    ('2025-09-22 00:37:45', '2025-09-22 00:37:55'),
    ('2025-09-22 00:38:10', '2025-09-22 00:38:20')
]
for start, end in active_intervals:
    df.loc[start:end, 'Label'] = 'active'

# กำหนดช่วงเวลาที่ 'rest'
rest_intervals = [
    ('2025-09-22 00:37:33', '2025-09-22 00:37:45'),
    ('2025-09-22 00:37:55', '2025-09-22 00:38:05')
]
for start, end in rest_intervals:
    df.loc[start:end, 'Label'] = 'rest'

# ลบข้อมูลที่ไม่ได้ถูกติดป้ายกำกับออก
df = df[df['Label'] != 'unknown'].copy()
print(f"\nจำนวนข้อมูลที่ถูกติดป้ายกำกับ: {len(df)}")
print(df['Label'].value_counts())

# 3. การสกัดคุณสมบัติและการเตรียมข้อมูลสำหรับโมเดล
window_size = 200
step_size = 50

X_features = []
y_labels = []

# ทำการ Windowing และสกัดคุณสมบัติ
for i in range(0, len(df) - window_size, step_size):
    window_data = df.iloc[i:i + window_size]
    # ตรวจสอบว่าทุกจุดใน window มี label เดียวกันหรือไม่
    if len(window_data['Label'].unique()) == 1:
        signal = window_data['Raw'].values
        
        # สกัดคุณสมบัติ
        mav = np.mean(np.abs(signal))
        zc = np.sum(np.diff(np.sign(signal)))
        wl = np.sum(np.abs(np.diff(signal)))
        
        X_features.append([mav, zc, wl])
        y_labels.append(window_data['Label'].iloc[0])

X = np.array(X_features)
y = np.array(y_labels)

if len(X) == 0:
    print("\nไม่พบข้อมูลที่ถูกติดป้ายกำกับในรูปแบบ windowed. โปรดตรวจสอบช่วงเวลาที่กำหนดอีกครั้ง")
else:
    # 4. การสร้างและฝึกโมเดล Random Forest
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
    rf_model.fit(X_train, y_train)

    # 5. การประเมินผล
    y_pred = rf_model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"\nความแม่นยำของโมเดล Random Forest: {accuracy*100:.2f}%")

    # พล็อต Confusion Matrix เพื่อดูผลลัพธ์อย่างละเอียด
    cm = confusion_matrix(y_test, y_pred, labels=['active', 'rest'])
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['active', 'rest'], yticklabels=['active', 'rest'])
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
```

```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# 1. การโหลดข้อมูลและสร้างป้ายกำกับด้วยมือ
df = pd.read_csv('emg_log2.csv')
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df.set_index('Timestamp', inplace=True)

# พล็อตข้อมูลเพื่อให้คุณเห็นภาพรวม
plt.figure(figsize=(15, 6))
plt.plot(df.index, df['Voltage'], label='Raw Voltage Signal', color='skyblue')
plt.title('EMG Signal from emg_log2.csv')
plt.xlabel('Timestamp')
plt.ylabel('Voltage')
plt.grid(True)
plt.legend()
plt.show()

# ************** ช่วงเวลาที่ปรับปรุงใหม่ให้ตรงกับไฟล์ emg_log2.csv **************
df['Label'] = 'unknown'

active_intervals = [
    # ช่วงเวลาที่มีสัญญาณสูง สันนิษฐานว่าเป็นการกำมือ
    ('2025-09-22 00:35:10', '2025-09-22 00:35:18'),
    ('2025-09-22 00:35:56', '2025-09-22 00:36:04'),
]
for start, end in active_intervals:
    df.loc[start:end, 'Label'] = 'active'

rest_intervals = [
    # ช่วงเวลาที่มีสัญญาณต่ำ สันนิษฐานว่าเป็นการคลายมือ
    ('2025-09-22 00:35:00', '2025-09-22 00:35:05'),
    ('2025-09-22 00:35:25', '2025-09-22 00:35:35'),
]
for start, end in rest_intervals:
    df.loc[start:end, 'Label'] = 'rest'

df = df[df['Label'] != 'unknown'].copy()
print(f"จำนวนข้อมูลที่ถูกติดป้ายกำกับ: {len(df)}")
print(df['Label'].value_counts())

# 2. การ Windowing และ Reshaping ข้อมูลสำหรับ CNN (โค้ดที่แก้ไขแล้ว)
window_size = 200
step_size = 50

X_windows = []
y_labels = []

for i in range(0, len(df) - window_size, step_size):
    window_data = df.iloc[i:i + window_size]
    signal = window_data['Voltage'].values
    label = window_data['Label'].iloc[-1]
    
    X_windows.append(signal)
    y_labels.append(label)

X = np.array(X_windows)
y = np.array(y_labels)

if X.ndim == 1:
    if len(X) > 0:
        X = X.reshape(-1, window_size, 1)
    else:
        X = None
elif X.ndim == 2:
    X = X.reshape((X.shape[0], X.shape[1], 1))

if X is not None and len(X) > 0:
    # 3. การเตรียม Label และแบ่งข้อมูล
    le = LabelEncoder()
    y_encoded = le.fit_transform(y)
    y_categorical = to_categorical(y_encoded)

    X_train, X_test, y_train, y_test = train_test_split(X, y_categorical, test_size=0.3, random_state=42)

    # 4. การสร้างโมเดล CNN
    model = Sequential([
        Conv1D(filters=64, kernel_size=5, activation='relu', input_shape=(window_size, 1)),
        MaxPooling1D(pool_size=2),
        Dropout(0.25),
        Flatten(),
        Dense(units=100, activation='relu'),
        Dense(units=y_categorical.shape[1], activation='softmax')
    ])

    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    model.summary()

    # 5. การฝึกโมเดลและประเมินผล
    history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2, verbose=1)

    y_pred_probs = model.predict(X_test)
    y_pred = np.argmax(y_pred_probs, axis=1)
    y_true = np.argmax(y_test, axis=1)
    
    accuracy = accuracy_score(y_true, y_pred)
    print(f"\nความแม่นยำบนข้อมูลทดสอบ: {accuracy*100:.2f}%")

    # พล็อต Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

    # พล็อต Training History
    plt.figure(figsize=(12, 6))
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid(True)
    plt.show()

else:
    print("ไม่สามารถสร้างและฝึกโมเดลได้เนื่องจากข้อมูล Windowed ไม่เพียงพอ")
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# โหลดสัญญาณ
df = pd.read_csv("emg_log3.csv")
signal = df["Voltage"].values

# สร้าง window
window_size = 200
step = 100
windows = []
for start in range(0, len(signal) - window_size, step):
    windows.append(signal[start:start+window_size])

windows = np.array(windows)

# ฟังก์ชันดึง feature
def extract_features(windows):
    feature_list = []
    for w in windows:
        mean_val = np.mean(w)
        range_val = np.max(w) - np.min(w)
        feature_list.append([mean_val, range_val])
    return np.array(feature_list)

features = extract_features(windows)

# Normalize feature 0-1
scaler = MinMaxScaler()
features_norm = scaler.fit_transform(features)

print("Feature shape:", features_norm.shape)
print("ตัวอย่าง feature (mean, range):")
print(features_norm[:5])
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# -------------------
# 1. โหลดข้อมูล
# -------------------
data = pd.read_csv("emg_log3.csv")

# สมมติใช้ column Voltage (ปรับเป็นชื่อคอลัมน์ของคุณ)
signal = data["Voltage"].values.reshape(-1, 1)

# Normalize ให้อยู่ในช่วง [0,1]
scaler = MinMaxScaler()
signal_norm = scaler.fit_transform(signal).flatten()

# -------------------
# 2. สร้าง window
# -------------------
window_size = 100  # กำหนดความยาว window
step = 50          # กำหนดการเลื่อน window

windows = []
features = []
for start in range(0, len(signal_norm) - window_size, step):
    w = signal_norm[start:start+window_size]
    mean_val = np.mean(w)
    range_val = np.max(w) - np.min(w)
    windows.append(w)
    features.append([mean_val, range_val])

features = np.array(features)

# -------------------
# 3. คัดเฉพาะ window ที่นิ่ง (range < 0.6)
# -------------------
stable_idx = np.where(features[:,1] < 0.6)[0]
print(f"เจอ window ที่นิ่ง {len(stable_idx)} อัน")

# -------------------
# 4. Plot ผลลัพธ์
# -------------------
plt.figure(figsize=(12, 6))
for i, idx in enumerate(stable_idx[:5]):  # plot แค่ 5 อันแรก
    start = idx * step
    w = windows[idx]
    mean_val = features[idx,0]
    plt.plot(range(start, start+window_size), w, label=f"Win {i} mean={mean_val:.2f}")
    
plt.title("Stable windows (range < 0.6)")
plt.xlabel("Sample")
plt.ylabel("Normalized EMG")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------
# 1. Load CSV
# ----------------------
data = pd.read_csv("emg_log3.csv")

# สร้าง digital_norm จาก Raw
data["digital_norm"] = data["Raw"] / 4095.0

# ----------------------
# 2. สร้าง window และ extract feature
# ----------------------
window_size = 50
windows = []
features = []

for start in range(0, len(data) - window_size, window_size):
    segment = data["digital_norm"].iloc[start:start+window_size].values
    mean_val = np.mean(segment)
    range_val = np.max(segment) - np.min(segment)
    windows.append(segment)
    features.append([mean_val, range_val])

features = np.array(features)

df_features = pd.DataFrame(features, columns=["mean", "range"])
print(df_features)

# ----------------------
# 3. เลือกช่วงที่นิ่งที่สุด (20% ล่างสุดของ range ;stable windows (range ต่ำสุด))
# ----------------------
percentile = 20
cutoff = np.percentile(features[:, 1], percentile)
print("cutoff (20%):", cutoff)

stable_idx = np.where(features[:, 1] <= cutoff)[0]
print("เจอช่วงนิ่ง:", len(stable_idx))

# ----------------------
# 4. วาดกราฟเฉพาะ stable windows
# ----------------------
plt.figure(figsize=(10, 5))

for idx in stable_idx:
    segment = windows[idx]
    mean_val = features[idx, 0]  # mean ของ window
    plt.plot(segment, alpha=0.7)  # ไม่ใส่ legend เยอะเกินไป

plt.title(f"Stable windows (lowest {percentile}% ranges)")
plt.xlabel("Sample")
plt.ylabel("Normalized EMG")
plt.show()
```
