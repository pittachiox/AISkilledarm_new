```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------
# 1. Load CSV
# ----------------------
data = pd.read_csv("emg_log1.csv").rename(columns={"Raw": "digital", "Voltage": "volt"})


# ----------------------
# 2. Preprocessing
# ----------------------
data["digital_norm"] = data["digital"] / 4095.0

# Moving Average Filter
window = 50
data["digital_filt"] = data["digital_norm"].rolling(window, center=True).mean()

# ----------------------
# 3. Threshold-based Labeling
# ----------------------
labels = []
for val in data["digital"]:
    if val > 2500:
        labels.append("grip")       #  
    elif val < 500:
        labels.append("open")       # 
    elif 1000 <= val <= 2000:
        labels.append("hold")       # 
    else:
        labels.append("transition") # 

data["label"] = labels

# ----------------------
# 4. Visualization
# ----------------------
plt.figure(figsize=(12,6))
plt.plot(data["digital"], label="Raw Digital")
plt.plot(data["digital_filt"]*4095, label="Filtered (scaled)")
plt.title("EMG Digital Signal with Grip/Open States")
plt.xlabel("Sample")
plt.ylabel("ADC value")
plt.legend()
plt.show()

# Show label distribution
print(data["label"].value_counts())
```

```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, LSTM, Dense, MaxPooling1D, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# 1. 

#  emg_log3.csv
df = pd.read_csv('emg_log3.csv')
emg_signal = df['Voltage'].values

#  window  window
window_size = 200  #  window
step_size = 50     # 

#  windowing  label  ( threshold-based)
X_list = []
y_list = []
threshold = 0.5  #  ()

for i in range(0, len(emg_signal) - window_size, step_size):
    window = emg_signal[i:i + window_size]
    #  (window_size, 1)  CNN
    X_list.append(window.reshape(-1, 1)) 
    
    #  window
    if np.mean(np.abs(window)) > threshold:
        y_list.append('active')
    else:
        y_list.append('rest')

#  list  numpy array
X = np.array(X_list)
y = np.array(y_list)

print(f'Shape  X (windowed data): {X.shape}')
print(f'Shape  y (labels): {y.shape}')

# 
unique, counts = np.unique(y, return_counts=True)
print(dict(zip(unique, counts)))

# 2. 

# Encode label
le = LabelEncoder()
y_enc = le.fit_transform(y)
y_cat = to_categorical(y_enc)

# 
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.2, random_state=42)

#  CNN-LSTM
model = Sequential([
    Conv1D(filters=64, kernel_size=5, activation='relu', input_shape=(window_size, 1)),
    MaxPooling1D(pool_size=2),
    Dropout(0.25),
    LSTM(units=100, return_sequences=False),
    Dense(units=len(np.unique(y)), activation='softmax')
])

# Compile model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# 
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.1, verbose=1)

# 3. 
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f': {accuracy*100:.2f}%')

# 
plt.figure(figsize=(12, 6))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 
df = pd.read_csv('emg_log3.csv')
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df.set_index('Timestamp', inplace=True)

# 
plt.figure(figsize=(15, 6))
plt.plot(df.index, df['Raw'], label='Raw Raw Signal', color='skyblue')
plt.title('EMG Signal from emg_log3.csv')
plt.xlabel('Timestamp')
plt.ylabel('Raw')
plt.grid(True)
plt.legend()
plt.show()



# 2.  (Manual Labeling)
# **************  **************
#  'unknown'
df['Label'] = 'unknown'

# :
#  'active'
active_intervals = [
    ('2025-09-22 00:37:45', '2025-09-22 00:37:55'),
    ('2025-09-22 00:38:10', '2025-09-22 00:38:20')
]
for start, end in active_intervals:
    df.loc[start:end, 'Label'] = 'active'

#  'rest'
rest_intervals = [
    ('2025-09-22 00:37:33', '2025-09-22 00:37:45'),
    ('2025-09-22 00:37:55', '2025-09-22 00:38:05')
]
for start, end in rest_intervals:
    df.loc[start:end, 'Label'] = 'rest'

# 
df = df[df['Label'] != 'unknown'].copy()
print(f"\n: {len(df)}")
print(df['Label'].value_counts())

# 3. 
window_size = 200
step_size = 50

X_features = []
y_labels = []

#  Windowing 
for i in range(0, len(df) - window_size, step_size):
    window_data = df.iloc[i:i + window_size]
    #  window  label 
    if len(window_data['Label'].unique()) == 1:
        signal = window_data['Raw'].values
        
        # 
        mav = np.mean(np.abs(signal))
        zc = np.sum(np.diff(np.sign(signal)))
        wl = np.sum(np.abs(np.diff(signal)))
        
        X_features.append([mav, zc, wl])
        y_labels.append(window_data['Label'].iloc[0])

X = np.array(X_features)
y = np.array(y_labels)

if len(X) == 0:
    print("\n windowed. ")
else:
    # 4.  Random Forest
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
    rf_model.fit(X_train, y_train)

    # 5. 
    y_pred = rf_model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"\n Random Forest: {accuracy*100:.2f}%")

    #  Confusion Matrix 
    cm = confusion_matrix(y_test, y_pred, labels=['active', 'rest'])
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['active', 'rest'], yticklabels=['active', 'rest'])
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
```

```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 
df = pd.read_csv('emg_log2.csv')
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df.set_index('Timestamp', inplace=True)

# 
plt.figure(figsize=(15, 6))
plt.plot(df.index, df['Voltage'], label='Raw Voltage Signal', color='skyblue')
plt.title('EMG Signal from emg_log2.csv')
plt.xlabel('Timestamp')
plt.ylabel('Voltage')
plt.grid(True)
plt.legend()
plt.show()

# **************  emg_log2.csv **************
df['Label'] = 'unknown'

active_intervals = [
    #  
    ('2025-09-22 00:35:10', '2025-09-22 00:35:18'),
    ('2025-09-22 00:35:56', '2025-09-22 00:36:04'),
]
for start, end in active_intervals:
    df.loc[start:end, 'Label'] = 'active'

rest_intervals = [
    #  
    ('2025-09-22 00:35:00', '2025-09-22 00:35:05'),
    ('2025-09-22 00:35:25', '2025-09-22 00:35:35'),
]
for start, end in rest_intervals:
    df.loc[start:end, 'Label'] = 'rest'

df = df[df['Label'] != 'unknown'].copy()
print(f": {len(df)}")
print(df['Label'].value_counts())

# 2.  Windowing  Reshaping  CNN ()
window_size = 200
step_size = 50

X_windows = []
y_labels = []

for i in range(0, len(df) - window_size, step_size):
    window_data = df.iloc[i:i + window_size]
    signal = window_data['Voltage'].values
    label = window_data['Label'].iloc[-1]
    
    X_windows.append(signal)
    y_labels.append(label)

X = np.array(X_windows)
y = np.array(y_labels)

if X.ndim == 1:
    if len(X) > 0:
        X = X.reshape(-1, window_size, 1)
    else:
        X = None
elif X.ndim == 2:
    X = X.reshape((X.shape[0], X.shape[1], 1))

if X is not None and len(X) > 0:
    # 3.  Label 
    le = LabelEncoder()
    y_encoded = le.fit_transform(y)
    y_categorical = to_categorical(y_encoded)

    X_train, X_test, y_train, y_test = train_test_split(X, y_categorical, test_size=0.3, random_state=42)

    # 4.  CNN
    model = Sequential([
        Conv1D(filters=64, kernel_size=5, activation='relu', input_shape=(window_size, 1)),
        MaxPooling1D(pool_size=2),
        Dropout(0.25),
        Flatten(),
        Dense(units=100, activation='relu'),
        Dense(units=y_categorical.shape[1], activation='softmax')
    ])

    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    model.summary()

    # 5. 
    history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2, verbose=1)

    y_pred_probs = model.predict(X_test)
    y_pred = np.argmax(y_pred_probs, axis=1)
    y_true = np.argmax(y_test, axis=1)
    
    accuracy = accuracy_score(y_true, y_pred)
    print(f"\n: {accuracy*100:.2f}%")

    #  Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

    #  Training History
    plt.figure(figsize=(12, 6))
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid(True)
    plt.show()

else:
    print(" Windowed ")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# -------------------
# 1. 
# -------------------
data = pd.read_csv("emg_log3.csv")

#  column Voltage ()
signal = data["Voltage"].values.reshape(-1, 1)

# Normalize  [0,1]
scaler = MinMaxScaler()
signal_norm = scaler.fit_transform(signal).flatten()

# -------------------
# 2.  window
# -------------------
window_size = 100  #  window
step = 50          #  window

windows = []
features = []
for start in range(0, len(signal_norm) - window_size, step):
    w = signal_norm[start:start+window_size]
    mean_val = np.mean(w)
    range_val = np.max(w) - np.min(w)
    windows.append(w)
    features.append([mean_val, range_val])

features = np.array(features)

# -------------------
# 3.  window  (range < 0.6)
# -------------------
stable_idx = np.where(features[:,1] < 0.6)[0]
print(f" window  {len(stable_idx)} ")

# -------------------
# 4. Plot 
# -------------------
plt.figure(figsize=(12, 6))
for i, idx in enumerate(stable_idx[:5]):  # plot  5 
    start = idx * step
    w = windows[idx]
    mean_val = features[idx,0]
    plt.plot(range(start, start+window_size), w, label=f"Win {i} mean={mean_val:.2f}")
    
plt.title("Stable windows (range < 0.6)")
plt.xlabel("Sample")
plt.ylabel("Normalized EMG")
plt.legend()
plt.show()
```
